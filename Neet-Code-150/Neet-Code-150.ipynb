# Q5 Top-K Elements

```python
nums = [1,4,4,5,5,5]
k = 2

def topKFrequent(nums, k):
    output = []
    values = {}
    for num in set(nums):
        values[num] = nums.count(num)
    
    vals = sorted(values.items(), key=lambda x: x[1])
    vals = vals[-k:]

    for val in vals:
        output.append(val[0])

    return output

topKFrequent(nums, k)
```

# Q4 Group Anagrams

```python
strs = ["act","pots","tops","cat","stop","hat"]
output = {}

def groupAnagrams(strs):
    for word in strs:
        key = ''.join(sorted(word))
    
        if key not in output:
            output[key] = []
        
        output[key].append(word)
    return list(output.values())

groupAnagrams(strs)
```
